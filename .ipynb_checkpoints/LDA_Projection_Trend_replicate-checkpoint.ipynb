```python
import os

import scipy.io
import scipy.stats as stats

import pandas as pd
import numpy as np 

from sklearn.model_selection import LeaveOneGroupOut, cross_val_score, cross_val_predict
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, roc_auc_score

from mat_preproc import preproc
```

## Data Reading

```python
all_data_paths = os.listdir("SN_vs_MN_data")
# I think those 12 files are the four duplicates of each of the 
# followings

# 9 features
fn400 = [paths for paths in all_data_paths if "fn400" in paths]
# 16 features
lpc = [paths for paths in all_data_paths if "lpc" in paths]
# 72 features
other = [paths for paths in all_data_paths if "lpc" not in paths and "fn400" not in paths]
```

Assuming each feature is disctint, we have 97 features in total

## LPC Features' fit

```python
# SN vs MN Data Preparation
# fn400
data_preproc = preproc(f"SN_vs_MN_data/{lpc[0]}")
pos1, neg1 = data_preproc.filter_index(2,5,2,4)
pos2, neg2 = data_preproc.filter_index(4,5,4,4)
pos_idx, neg_idx = data_preproc.merge_two_class(pos1, neg1, pos2, neg2)
X, y, subject = data_preproc.get_data_by_index(pos_idx, neg_idx)

logo = LeaveOneGroupOut()

print(f"there are {logo.get_n_splits(X, y, subject)} groups")
```

```python
# positive class percentage
y.mean()
```

```python
# Evaluate within fold
logo = LeaveOneGroupOut()
LDA = LinearDiscriminantAnalysis(shrinkage = 'auto', solver = 'eigen')
scores = cross_val_score(LDA, X, y, cv=logo, groups=subject)
scores.mean()
```

```python
# Evaluate across fold
LDA = LinearDiscriminantAnalysis(shrinkage = "auto", solver = 'eigen')
pred = cross_val_predict(LDA, X, y, cv=logo, groups=subject)
accuracy_score(y, pred)
```

# Combined All Features

```python
# SN vs MN Data Preparation
# combined all feature. They are all according to the same sequence of y
all_data = [other[0], fn400[0], lpc[0]]

X_all = np.array([])
y_all = None

for filepath in all_data:
    data_preproc = preproc(f"SN_vs_MN_data/{filepath}")
    pos1, neg1 = data_preproc.filter_index(2,5,2,4)
    pos2, neg2 = data_preproc.filter_index(4,5,4,4)
    pos_idx, neg_idx = data_preproc.merge_two_class(pos1, neg1, pos2, neg2)
    X, y, subject = data_preproc.get_data_by_index(pos_idx, neg_idx)
    try:
        X_all = np.hstack([X_all, X])
    except ValueError:
        X_all = X
    y_all = y
```

```python
logo = LeaveOneGroupOut()
print(f"there are {logo.get_n_splits(X_fn400, y_fn400, subject)} groups")
LDA = LinearDiscriminantAnalysis(shrinkage = 'auto', solver = 'lsqr')
scores = cross_val_score(LDA, X_all, y_all, cv=logo, groups=subject)
scores.mean()
```

```python
LDA = LinearDiscriminantAnalysis(shrinkage = 1, solver = 'lsqr')
pred = cross_val_predict(LDA, X_all, y_all, cv=logo, groups=subject)
accuracy_score(y, pred)
```
